```python
def get_avg(input_nos):
    sum =0.0
    for num in input_nos:
        sum +=num

    avg= sum/len(input_nos)
    return avg
```

```python
input_nos=[5.0,3.5,7.8,9.9,10.0]
avg=get_avg(input_nos)
print('The average is:',avg)
```

```python
if avg > 5.0:
    print('The avaerage is too high!')
```

```python
# checks first and last element is equal or not in the list

def is_first_last_equal(nos_list):
    if len(nos_list) == 0:
        return
    if nos_list[0] == nos_list[-1]:
        return True
    else:
        return False
```

```python
print(is_first_last_equal([10,20,30,10]))
```

```python
print(is_first_last_equal([10,20,30,40]))
```

```python
print(is_first_last_equal([]))
```

```python
# bitwise AND
first_bit = 1
second_bit = 1

print(first_bit & second_bit)
```

```python
# bitwise OR
first_bit = 0
second_bit = 1

print(first_bit | second_bit)
```

```python
# bitwise ^ XOR
first_bit = 1
second_bit = 1

print(first_bit ^ second_bit)
```

```python
# bitwise NOT ~x = -x -1

print(~2)
```

```python
# However, the in operator can also be used to check whether a
# given element is present in a list

invited_guests = ['Prashant','Gurudatta','Nandu','Pratik','Gopal']
name = input('What\'s your name? ')

if name in invited_guests:
    print('Welcome!')
else:
    print('You are not invited!')
    
```

```python
# similary, we can use a negated in operator, not in.

invited_guests = ['Prashant','Gurudatta','Nandu','Pratik','Gopal']
name = input('What\'s your name? ')

if name not in invited_guests:
    print('You are not invited!')
else:
    print('Welcome')
    
```

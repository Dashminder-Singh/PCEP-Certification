```python
# Dictionaries are the collection to store key-value pairs

emails ={
        'Anne Stahl' : 'astahl@gmail.com',
        'Peter Small' : 'peters@yandex.com',
        'Mark Steel' : 'mark@steel.com'
}
```

```python
print(emails['Mark Steel'])
# emails['Mark Steel']
```

```python
for i in range(10):  # by default start with zero
    print(i)
print('Finished!')
```

```python
for i in range(1,11):  # start with inclusive and stop is exclusive
    print(i)
print('Finished!')
```

```python
while True: # infine loop - always true
    name = input('Enter a name or EXIT to close the program: ')
    if name == 'EXIT':
        break
    print('Hello ',name)
```

```python
for i in range(1,21):
    if i%5 ==0:
        continue
    print(i)
```

```python
for i in range (1,6): #Nested loops
    for j in range (1,6):
        print(i,'X',j,'=',i*j)
```

```python
i=1 # The else branch of a loop is always executed exaclty once 
    #( exception: break statement). The only time the else branch is not exceuted
    # when a break statement is executed inside the loops

while i< 5:
    print(i)
    i+=1
else:
    print('else:',i)
```

```python
guess_year = 1994

while True:
    user_input =int(input('When was Python 1.0 released? '))
    if user_input == guess_year:
        print('Correct')
        break
    elif user_input > guess_year:
        print('It was earlier than that!')
    else:
        print('It was later than that!')
```

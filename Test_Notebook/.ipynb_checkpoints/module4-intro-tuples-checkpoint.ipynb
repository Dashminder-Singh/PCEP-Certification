```python
empty_tuple =()
print(empty_tuple)
```

```python
one_el_tuple_a = (1,)
print(one_el_tuple_a)
```

```python
# note, we had to introduce a comma after the element. When
# you have more than one value in your tuple, the final comma is optional
one_el_tuple_b = 1,
print(one_el_tuple_b)
```

```python
# three_el_tuple = 1,2,3,
three_el_tuple = 1, 2, 3
print(three_el_tuple)
```

```python
# mutable vs Immutable
# mutable data can be freely updated anytime you want.
# example, List
# But in case of tuple, once it's created - it stays this way forever
# But we can assign a new tuple to the same variable

user_data = ('John','American',1964)
user_data =('James','American',1980)
print(user_data)
```

```python
user_data.append('Horshe')
```

```python
del user_data[0]
```

```python
print(user_data[0])
```

```python
user_data[0] ='Mark'
```

```python
# Incidentally, strings in python are also immutable
message ='welcome'
message ='Welcome'
```

```python
print(message)
```

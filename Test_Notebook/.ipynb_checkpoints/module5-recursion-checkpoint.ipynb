```python
# factorilal
# 3! = 1 * 2 * 3 = 6
# 5! = 1 * 2 *3 *4 *5 = 120
```

```python
# iterative
def get_factorial(no):
    fact = 1
    for i in range(1,no+1):
        fact *= i
    return fact
```

```python
print(get_factorial(5))
```

```python
# 1! = 1
# 2! = 2 * 1!
# 3! = 3 * 2!
# 4! = 4 * 3!
# 5! = 5 * 4!
```

```python
# recursive

def get_factorial_recursive(no):
    if no <= 1:
        return 1
    return no * get_factorial_recursive(no-1)
    
```

```python
print(get_factorial_recursive(3))
```

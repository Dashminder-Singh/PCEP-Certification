```python
grades = {}
grades['John'] ='A'
grades['Anne'] ='B'

for el in grades:
    print(el)
```

```python
for el in grades.keys():
    print(el)
```

```python
for el in grades.values():
    print(el)
```

```python
for person, grade in grades.items():
    print(person,'got',grade)
```

```python
sample_dict = {
    "mouth": "Mund",
    "finger": "Finger",
    "leg": "Bein",
    "hand": "Hand",
    "face": "Gesicht",
    "nose": "Nase"
}
```

```python
print(sample_dict['face'])
```

```python
sample_dict.items()
```

```python
while True:
    
    user= input('Enter a word in English or EXIT: ')
    
    if user in sample_dict:
        print('Translation:',sample_dict[user])
    elif user == 'EXIT':
        break
    else:
        print('No match!')
print('Bye!')
```

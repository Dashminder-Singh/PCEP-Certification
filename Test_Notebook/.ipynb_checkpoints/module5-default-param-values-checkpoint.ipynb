```python
#keyword arguments
# default parameters
def count_letter(text='Letters',letter='e'):
    count=0
    for ch in text:
        if ch==letter:
            count +=1
    print('Number of',letter,'is',count)
    
```

```python
count_letter('Search')
```

```python
count_letter()
```

```python
# ALl positional arguments must apprear before any named arguments
count_letter('Search here', letter='r')
```

```python
# remove the duplicates elements  and return a new list

def unique(nos_list=[]):
    temp_list=[]
    
    for no in nos_list:
        if no not in temp_list:
            temp_list.append(no)
    return temp_list
```

```python
print(unique([10,20]))
```

```python
print(unique([2,1,1,2,4,6,6,]))
```

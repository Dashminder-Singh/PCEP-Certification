```python
name_original = 'John Snow'
name_new = name_original
name_original = 'Daenerys Targaryen'

print('Orginal:',name_original,'New:',name_new)

# you can see that we updated name original, but did not
# change name in any way.The same rule applies to other basic
# data types integers, float, and booleans
```

```python
# However, the same rule is not true for data collections like lists
list_original = [1,2,3]
list_new = list_original
list_original[0] = 10

print('Orginal:',list_original,'New:',list_new)
```

```python
# If we now assign the value of the original variable to the
# new variable, all we do is we copy the reference to the original
# list behind the scenes.
# Both variables now point to the very same place in the memory with
# the very same list.
```

```python
# To modify a list indentpentdly, we use the list original slicing
list_original = [1,2,3]
list_new = list_original[:]
list_original[0] = 10

print('Orginal:',list_original,'\nNew:',list_new)
```

```python
list_original = [1,2,3]
list_new = list_original[:2]
list_original[0] = 10

print('Orginal:',list_original,'\nNew:',list_new)
```

```python
def get_no():
    for i in range(1,4):
        yield i
```

```python
print(get_no())
```

```python
gen = get_no()
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
for i in get_no():
    print(i)
```

```python
nos = list(get_no()) # turned into a list
```

```python
input_nos = [5.0, 3.5, 7.8, 9.9, 10.0]

def get_avg(input_no): #local parameter
    sum = 0.0
    for no in input_no:
        sum +=no
    avg = sum/len(input_no)
    print(avg)

get_avg(input_nos) #passing a argument(list)
```

```python
get_avg()
```

```python
get_avg(5)
```

```python
# count the letter in a give string or text

def count_letter(text,letter):
    count=0
    for ch in text:
        if ch==letter:
            count +=1

    print('Number of',letter,'is',count)

count_letter('MadaM','M')
```

```python
count_letter('People say nothing is impossible, but I do nothing every day.','e')
```

```python
count_letter('e','Welcome') # the order of arugument is important here
```

```python
# The argument values are assigned to the parameters based on the position.

count_letter(text='aadaM',letter='a')
```

```python
count_letter(letter='a', text='aadaM')
```

```python
connections = [
    ('Amsterdam', 'Dublin', 100),
    ('Amsterdam', 'Rome', 140),
    ('Rome', 'Warsaw', 130),
    ('Minsk', 'Prague', 95),
    ('Stockholm', 'Rome', 190),
    ('Copenhagen', 'Paris', 120),
    ('Madrid', 'Rome', 135),
    ('Lisbon', 'Rome', 170),
    ('Dublin', 'Rome', 170),
    ]
count =0
add =0
for item in connections:
    if item[1] =='Rome':
        count +=1
        add += item[2]

print(count,'connections lead to Rome with an average flight time of',add/count,'minutes')
```

```python
numbers = [1,2,3,4]
```

```python
numbers = []
for i in range (1,101):
    numbers.append(i)
print(numbers)
```

```python
# list comprehension -shortcut

numbers = [i  for i in range(1,101)]
print(numbers)
```

```python
# if you want to skip the numbers which is divide by 3

numbers = [i for i in range(1,101) if i%3 != 0]
print(numbers)
```

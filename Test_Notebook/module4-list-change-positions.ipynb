```python
first =input('Enter first number: ')
second = input('Enter second number: ')

print('Before swapping:',first,second)
first = second
second = first

print('After swapping:',first,second)
```

```python
first =input('Enter first number: ')
second = input('Enter second number: ')

print('Before swapping:',first,second)
temp = first
first = second
second = temp

print('After swapping:',first,second)
```

```python
first =input('Enter first number: ')
second = input('Enter second number: ')

print('Before swapping:',first,second)

first, second = second, first
print('After swapping:',first,second)
```
